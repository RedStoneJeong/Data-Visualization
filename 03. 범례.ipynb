# 03. 범례

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'HYHeadLine-Medium' #window
matplotlib.rcParams['font.size'] = 15 #글씨 사이즈
matplotlib.rcParams['axes.unicode_minus'] = False #한글폰트사용시 깨지는 -부호 
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x, y, label = '무슨데이터')
plt.legend()
```

```python
plt.plot(x, y, label = '무슨데이터')
plt.legend(loc='upper right')
```

```python
plt.plot(x, y, label = '무슨데이터')
plt.legend(loc='lower right')
```

```python
plt.plot(x, y, label = '범례')
plt.legend(loc=(0.7, 0.8)) # x, y축 기준으로 (0~1사이)
```
