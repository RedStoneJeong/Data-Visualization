# 02. 축

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'HYHeadLine-Medium' #window
matplotlib.rcParams['font.size'] = 15 #글씨 사이즈
matplotlib.rcParams['axes.unicode_minus'] = False #한글폰트사용시 깨지는 -부호 
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x, y)
plt.title('꺾은선 그래프', fontdict={'family' : 'HYGungSo-Bold', 'size':20})
```

```python
plt.plot(x,y)
plt.xlabel('x축')
plt.ylabel('y축')
```

```python
plt.plot(x, y)
plt.xlabel('x축', color='red')
plt.ylabel('y축', color='#00aa00')
```

```python
plt.plot(x,y)
plt.xlabel('x축', color='red', loc='right') #left, center, light
plt.ylabel('y축', color='#00aa00', loc='top') # top, center, bottom
```

```python
plt.plot(x,y)
plt.xticks([1,2,3])
plt.yticks([3,6,9,12])
plt.show()
```
