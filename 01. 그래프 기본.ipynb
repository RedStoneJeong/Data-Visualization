# matplotlib
다양한 형태의 그래프를 통해서 시각화를 할 수 있는 라이브러리

```python
import matplotlib.pyplot as plt
```

## 그래프 기본

```python
x = [1, 2, 3]
y = [2, 4, 8]
plt.plot(x, y)
```

```python
print(plt.plot(x,y))
# 동작의 리턴 결과 문구가 나타남
```

```python
plt.plot(x,y)
plt.show()
```

## title 설정

```python
plt.plot(x,y)
plt.title('Line Graph')
```

```python
plt.plot(x, y)
plt.title('꺾은선 그래프')
```

## 한글 폰트 설정

```python
import matplotlib
matplotlib.rcParams['font.family'] = 'HYHeadLine-Medium' #window
matplotlib.rcParams['font.size'] = 15 #글씨 사이즈
matplotlib.rcParams['axes.unicode_minus'] = False #한글폰트사용시 깨지는 -부호 
```

```python
import matplotlib.font_manager as fm
fm.fontManager.ttflist
[f.name for f in fm.fontManager.ttflist]
```
